```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Market
MIN_PRICE_APR_10 = 4050
MAX_PRICE_APR_10 = 4100
MIN_PRICE_APR_13 = 4100
MAX_PRICE_APR_13 = 4150
MIN_PRICE_APR_14 = 4100
MAX_PRICE_APR_14 = 4150
```

```python
# Read trading events
events = pd.read_csv('events.csv')
```

```python
# Plot AdjPnl and SP_Price
fig, ax1 = plt.subplots()
ax1.plot(events["ID"], events["AdjPnL"], color="tab:blue")

ax2 = ax1.twinx()
ax2.plot(events["ID"], events["SP_Price"], color="tab:red")
ax2.axhline(y=MIN_PRICE_APR_14, color='g', linestyle='-')
ax2.axhline(y=MAX_PRICE_APR_14, color='g', linestyle='-')
ax2.axes.get_yaxis().set_visible(False)

ax3 = ax1.twinx() 
ax3.plot(events["ID"], events["Position"], color="tab:orange")

ax1.legend(["Adjusted PnL"], bbox_to_anchor=(1.10, 1), loc="upper left")
ax2.legend(["S&P 500"], bbox_to_anchor=(1.10, 0.9), loc="upper left")
ax3.legend(["Position (yes - no)"], bbox_to_anchor=(1.10, 0.8), loc="upper left")
plt.title("Sample Simulation Results")
```
